```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['Machine\nLearning', 'Big Data', 'Business\nAnalytics', 'Business\nIntelligence', 'Databases']
fracs = np.full((5), 20)

plt.figure(figsize=(2.5, 2.5))
patches, texts, autotexts = plt.pie(fracs, labels=labels, radius = 1, startangle = 0, autopct='%1.1f%%', shadow = False, explode = np.full((5), 0.01), labeldistance = 0.6, textprops={'fontsize': 13})
plt.axis("equal")
plt.tight_layout()

[x.set_horizontalalignment('center') for x in texts]
[x.remove() for x in autotexts]
[x.set_color('w') for x in texts]
[x.set_size(9) for x in texts]
[x.set_weight('bold') for x in texts]
[x.set_color('w') for x in autotexts]

plt.show()
```
